## Vector Set
Collection  of  vectors  is  called  a  <b>set</b>.You  can  imagine  putting  a  bunch  of  vectors  into  a  bag  to  carry  around.  Vector  setsare indicated using capital italics letters, like S or V. Mathematically, we can describesets as the following:
V = {v1,v2,........,vn}.

V = {} means empty vector set is empty.


## Linear Weighted Combination
A linear weighted combination is a way of mixing information from multiple variables,with  some  variables  contributing  more  than  others.  This  fundamental  operationis  also  sometimes  called  linear  mixture  or  weighted  combination  (the  linear  part  isassumed). Sometimes, the term coefficient is used instead of weight.Linear  weighted  combination  simply  means  scalar-vector  multiplication  and  addi‐tion:  take  some  set  of  vectors,  multiply  each  vector  by  a  scalar,  and  add  them  toproduce a single vector.





<a id="linear_weighted_combination"></a>

```python
#Example for Linear Weighted Combination
lambda_1 = 1
lambda_2 = 2
lambda_3 = -3
v1 = np.array([[4,5,1]]).T
v2 = np.array([[-4,0,-4]]).T
v3 = np.array([[1,3,2]]).T
print(v1 , v2, v3)
w = lambda_1*v1 + lambda_2*v2 + lambda_3*v3
print(w)
```

## Linear Independence
A set of vectors is linearly dependent if at least one vector in the set can be expressedas  a  linear  weighted  combination  of  other  vectors  in  that  set.  And  thus,  a  set  of vectors  is  linearly  independent  if  no  vector  can  be  expressed  as  a  linear  weighted combination of other vectors in the set.
Independence is a property of a set of vectors. That is, a set of vectors can be linearly independent  or  linearly  dependent;  independence  is  not  a  property  of  an  individual vector within a set.



## Subspace and Span
For  some  (finite)  set  of  vectors,  the  infinite  number  of  ways  to  linearlycombine them—using the same vectors but different numerical values for the weights—creates  a  vector  subspace.  And  the  mechanism  of  combining  all  possible  linearweighted  combinations  is  called  the  span  of  the  vector  set.

In other words, The span of a set of vectors is simply the set of points that can be reached by walking in the direction of these vectors in any combination.

A group/set of vectors is said to be linearly independent if <b>NONE</b> of the vectors in that set can be obtained as a linear combination of others vectors present in that set.

```python
#Example 1
# Define the vector
V = np.array([1, 3])

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the vector V starting from origin in red with tail and head
ax.arrow(0, 0, V[0], V[1], head_width=0.1, head_length=0.2, fc='r', ec='r')

# Define the subspace line as a scalar multiple of the vector V
t = np.linspace(-5, 5, 400)  # Adjust the range of t for zoomed view
line_points = np.outer(t, V)

# Plot the subspace line in gray as a dashed line
ax.plot(line_points[:, 0], line_points[:, 1], '--', color='gray')

# Set the limits of the plot to zoom in on the vector
ax.set_xlim(-2, 2)  # Change the limits for zoomed view
ax.set_ylim(-2, 6)  # Change the limits for zoomed view

# Add grid
ax.grid()

# Set the aspect ratio of the plot to be equal
ax.set_aspect('equal')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Vector Plot of V = [1, 3] and Its Subspace')

# Show the plot
plt.show()

```

The  span  of  this  vector  set  is  the  infinity  of  vectors  that  can  be  created  as  linear combinations of the vectors in the set. For a set with one vector, that simply means all possible scaled versions of that vector.
<b>The vector is in red and the subspace it spans is grey</b>


```python
# Another example 2

V1 = np.array([1, 0, 2])
V2 = np.array([-1, 1, 2])

# Create a figure and axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the vector V1 starting from origin in red with tail and head
ax.quiver(0, 0, 0, V1[0], V1[1], V1[2], color='r', arrow_length_ratio=0.1, label='Vector V1')

# Plot the vector V2 starting from origin in blue with tail and head
ax.quiver(0, 0, 0, V2[0], V2[1], V2[2], color='b', arrow_length_ratio=0.1, label='Vector V2')

# Define the subspace lines as scalar multiples of the vectors V1 and V2
t = np.linspace(-5, 5, 400)
line_points_V1 = np.outer(t, V1)
line_points_V2 = np.outer(t, V2)

# Plot the subspace lines in gray as dashed lines
ax.plot(line_points_V1[:, 0], line_points_V1[:, 1], line_points_V1[:, 2], '--', color='gray')
ax.plot(line_points_V2[:, 0], line_points_V2[:, 1], line_points_V2[:, 2], '--', color='gray')

# Set the limits of the plot to zoom in on the vectors
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 6])

# Add grid
ax.grid()

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Vector Plot of V = {[1,0,2],[-1,1,2]} and Their Subspaces')
ax.legend()

# Show the plot
plt.show()
```

The above example is a set of 2 vectors each in $R^3$ and hence they are represented in a 3-D axis. But the subspace that they span is a 2D Plane in that 3D space. That plane passes through the origin because scaling both the vectors by 0 gives the zero vectors.

<a id="subspace_example_cell_3"></a>

```python
#Example 3
V1 = np.array([1, 1, 1])
V2 = np.array([2, 2, 2])

# Create a figure and axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the vector V1 starting from origin in red with tail and head
ax.quiver(0, 0, 0, V1[0], V1[1], V1[2], color='r', arrow_length_ratio=0.3, label='Vector V1')

# Plot the vector V2 starting from origin in blue with tail and head
ax.quiver(0, 0, 0, V2[0], V2[1], V2[2], color='b', arrow_length_ratio=0.1, label='Vector V2')

# Define the subspace lines as scalar multiples of the vectors V1 and V2
t = np.linspace(-5, 5, 400)
line_points_V1 = np.outer(t, V1)
line_points_V2 = np.outer(t, V2)

# Plot the subspace lines in gray as dashed lines
ax.plot(line_points_V1[:, 0], line_points_V1[:, 1], line_points_V1[:, 2], '--', color='gray')
ax.plot(line_points_V2[:, 0], line_points_V2[:, 1], line_points_V2[:, 2], '--', color='gray')

# Set the limits of the plot to zoom in on the vectors
ax.set_xlim([-1, 4])
ax.set_ylim([-1, 4])
ax.set_zlim([-1, 6])

# Add grid
ax.grid()

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Vector Plot of V = {[1,0,2],[-1,1,2]} and Their Subspaces')
ax.legend()

# Show the plot
plt.show()
```

The  first  example  had  one  vector  and  its  span  was  a  1D  subspace,  and  the  second example  had  two  vectors  and  their  span was  a  2D  subspace.
But in the third example 2 example in $R^3$, but the subspace that they span is still only a 1D subspace —a line. It’s  because  one  vector  in  the  set  is  already  in  the  span  of the other vector. Thus, in terms of span, one of the two vectors is redundant.

#### The dimensionality of the subspace spanned by a set of vectors is the smallest number of vectors that forms a linearly independent set. If a vector set is linearly independent,then  the  dimensionality  of  the  subspace  spanned  by  the  vectors  in  that  set  equals the number  of  vectors  in  that  set.  If  the  set  is  dependent,  then  the  dimensionality  of the subspace spanned by those vectors is necessarily less than the number of vectors in  that  set.

### The  formal  definition  of  a  vector  subspace  is  a  subset  that  is  closed  under  addition and  scalar  multiplication,  and  includes  the  origin  of  the  space.  That  means  that  any linear  weighted  combination  of  vectors  in  the  subspace  must  also  be  in  the  same subspace,  including  setting  all  weights  to  zero  to  produce  the  zeros  vector  at  the origin of the space.

## Subspace vs Span -- Same or Different
Span and Sub space are highly related concepts and often refer to the same thing.Span and subspace so often refer to identical mathematical objects that using the terms interchangeably is usually correct.
Thinking  of  span  as  a  verb  and  subspace  as  a  noun  helps  understand  their distinction: a set of vectors spans, and the result of their spanning is a subspace. Now consider  that  a  [subspace  can  be  a  smaller  portion  of  a  larger  space , example 3](#subspace_example_cell_3) . 

Putting these together: span is the mechanism of creating a subspace. (On the  other  hand,  when  you  use  span  as  a  noun,  then  span and  subspace  refer  to  the same infinite vector set.)

## Basis
1. A basis is like a ruler for measuring a space.
2. A basis  is  a  set  of  rulers  that we  use  to  describe  the  information in the matrix (e.g., data).
3. A basis is a set of vectors that :-
    i. Spans a vector space.
    ii. Is linearly independent.
4. Basis are any 2 vectors (coming from origin) with the help of which one can reach to any point in the vector space.
5. Anything that comprises 2 vectors that go in the same direction or in the opposite direction but belong to the same line then they are not forming a basis (They are not linearly independent).
6. The length of the basis, is the space of the dimension of the plane.
7. The length of the basis, of a space is the dimension of that space.

### Basis is simply the combination of span and independence: a set of vectors can be a basis for some subspace if it (1) spans that subspace and (2) is an independent set of vectors.

### Question 3-1.Rewrite the code for linear weighted combination, but put the scalars in a list and the vectors  as  elements  in  a  list  (thus,  you  will  have  two  lists,  one  of  scalars  and  one  of NumPy  arrays).  Then  use  a  for  loop  to  implement  the  linear  weighted  combination operation.  Initialize  the  output  vector  using  np.zeros().  Confirm  that  you  get  the same result as in the [previous code](#linear_weighted_combination).

```python
lambda_list = [lambda_1,lambda_2,lambda_3]
vector_list = [v1,v2,v3]
output_vector = np.zeros(v1.shape)
for lambdas,vectors in zip(lambda_list,vector_list):
    output_vector = output_vector + lambdas*vectors
print(output_vector)
```
