# What is a Matrix?
A matrix is a vector taken to the next level. Matrices are highly versatile mathematical objects.  They  can  store  sets  of  equations,  geometric  transformations,  the  positions of  particles  over  time,  financial  records,  and  myriad  other  things.  In  data  science,matrices are sometimes called data tables, in which rows correspond to observations(e.g., customers) and columns correspond to features (e.g., purchases).
Depending on the context, matrices can be conceptualized as a set of column vectors stacked  next  to  each  other  (e.g.,  a  data  table  of  observations-by-features),  as  a  set  of row vectors layered on top of each other (e.g., multisensor data in which each row is a time series from a different channel), or as an ordered collection of individual matrix elements (e.g., an image in which each matrix element encodes pixel intensity value)
#### Matrices  are  indicated  using  bold-faced  capital  letters,  like  matrix  A  or  M.  The  size of  a  matrix  is  indicated  using  (row,  column)  convention.

## Special Matrix
Matrices can be described using a relatively small  number  of  characteristics,  which  creates  “families”  or  categories  of  matrices. These categories are important to know, because they appear in certain operations or have certain useful properties. Some  categories  of  matrices  are  used  so  frequently  that  they  have  dedicated  NumPy functions  to  create  them.  Following  is  a  list  of  some  common  special  matrices.:-
1. <i>Random numbers matrix</i>:- This  is  a  matrix  that  contains  numbers  drawn  at  random  from  some  distribution, typically Gaussian (a.k.a. normal). Random-numbers matrices are great for exploring linear algebra in code, because they are quickly and easily created with any size and rank.
```python
rows = 4
columns=6
A = np.random.randn(rows,columns)
```
2. <i>Square versus nonsquare</i> :- A square  matrix  has  the  same  number  of  rows  as  columns;  in  other  words,  the matrix  is  in $\mathbb{R}^{N \times N}$.  A  nonsquare  matrix,  also  sometimes  called  a  rectangular matrix,  has  a  different  number  of  rows  and  columns.  You  can  create  square  and rectangular matrices from random numbers by adjusting the shape parameters in the previous code. Rectangular  matrices  are  called  <b><i><u>tall</b></i></u>  if  they  have  more  rows  than  columns  and <b><i><u>wide</b></i></u> if they have more columns than rows.
3. <i>Diagonal</i>:- The  diagonal  of  a  matrix  is  the  elements  starting  at  the  top-left  and  going  down to the bottom-right. A diagonal matrix has zeros on all the off-diagonal elements; the diagonal elements may also contain zeros, but they are the only elements that may contain nonzero values.<br> The  NumPy  function  <b>np.diag()</b>  has  two  behaviors  depending  on  the  inputs: input  a  matrix  and  <b>np.diag</b>  will  return  the  diagonal  elements  as  a  vector;  input a  vector  and  <b>np.diag</b>  will  return  a  matrix  with  those  vector  elements  on  the diagonal. (Note: extracting the diagonal elements of a matrix is not called “diagonalizing a matrix”;
4. <i>Triangular</i>:- A  triangular  matrix  contains  all  zeros  either  above  or  below  the  main  diagonal.The  matrix  is  called  upper  triangular  if  the  non-zero  elements  are  above  the diagonal and lower triangular if the non-zero elements are below the diagonal.<br>NumPy  has  dedicated  functions  to  extract  the  <b>upper  (np.triu())  or  lower (np.tril()) triangle <b>of a matrix.
5. <i>Identity</i>:-The identity matrix is one of the most important special matrices. It is the equivalent of the number 1, in that any matrix or vector times the identity matrix is that same  matrix  or  vector.  The  identity  matrix  is  a  square  diagonal  matrix  with all diagonal elements having a value of 1. It is indicated using the letter <b>I</b>. You might see a subscript to indicate its size (e.g., I$_{5}$ is the 5 × 5 identity matrix); if not, then you can infer the size from context (e.g., to make the equation consistent).<br>You can create an identity matrix in Python using <b>np.eye()</b>.
6. <i>Zeros</i>:- The  zeros  matrix  is  comparable  to  the  zeros  vector:  it  is  the  matrix  of  all  zeros. Like  the  zeros vector,  it  is  indicated  using  a  bold-faced  zero:  <b>0</b>.  It  can  be  a  bit confusing to have the same symbol indicate both a vector and a matrix, but this kind of overloading is common in math and science notation.<br>The zeros matrix is created using the <b>np.zeros()</b> function

````python
# Define the size of the matrices
N = 5

# Create the matrices with random values
square_matrix = np.random.randn(N, N)  # Random square matrix
upper_triangle_matrix = np.triu(np.random.randn(N, N))  # Upper triangle matrix
lower_triangle_matrix = np.tril(np.random.randn(N, N))  # Lower triangle matrix
diagonal_matrix = np.diag(np.random.randn(N))  # Diagonal matrix
identity_matrix = np.eye(N)  # Identity matrix
zeros_matrix = np.zeros((N, N))  # Zeros matrix

# Plotting the matrices with annotations
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Plot square matrix
im = axs[0, 0].imshow(square_matrix, cmap='Greys', interpolation='nearest')
for i in range(N):
    for j in range(N):
        text = axs[0, 0].text(j, i, "{:.2f}".format(square_matrix[i, j]),
                               ha="center", va="center", color="black")

axs[0, 0].set_title('Square Matrix')
fig.colorbar(im, ax=axs[0, 0])

# Plot upper triangle matrix
im = axs[0, 1].imshow(upper_triangle_matrix, cmap='Greys', interpolation='nearest')
for i in range(N):
    for j in range(N):
        text = axs[0, 1].text(j, i, "{:.2f}".format(upper_triangle_matrix[i, j]),
                               ha="center", va="center", color="black")

axs[0, 1].set_title('Upper Triangle Matrix')
fig.colorbar(im, ax=axs[0, 1])

# Plot lower triangle matrix
im = axs[0, 2].imshow(lower_triangle_matrix, cmap='Greys', interpolation='nearest')
for i in range(N):
    for j in range(N):
        text = axs[0, 2].text(j, i, "{:.2f}".format(lower_triangle_matrix[i, j]),
                               ha="center", va="center", color="black")

axs[0, 2].set_title('Lower Triangle Matrix')
fig.colorbar(im, ax=axs[0, 2])

# Plot diagonal matrix
im = axs[1, 0].imshow(diagonal_matrix, cmap='Greys', interpolation='nearest')
for i in range(N):
    for j in range(N):
        text = axs[1, 0].text(j, i, "{:.2f}".format(diagonal_matrix[i, j]),
                               ha="center", va="center", color="black")

axs[1, 0].set_title('Diagonal Matrix')
fig.colorbar(im, ax=axs[1, 0])

# Plot identity matrix
im = axs[1, 1].imshow(identity_matrix, cmap='Greys', interpolation='nearest')
for i in range(N):
    for j in range(N):
        text = axs[1, 1].text(j, i, "{:.2f}".format(identity_matrix[i, j]),
                               ha="center", va="center", color="black")

axs[1, 1].set_title('Identity Matrix')
fig.colorbar(im, ax=axs[1, 1])

# Plot zeros matrix
im = axs[1, 2].imshow(zeros_matrix, cmap='Greys', interpolation='nearest')
for i in range(N):
    for j in range(N):
        text = axs[1, 2].text(j, i, "{:.2f}".format(zeros_matrix[i, j]),
                               ha="center", va="center", color="black")

axs[1, 2].set_title('Zeros Matrix')
fig.colorbar(im, ax=axs[1, 2])

# Adjust layout
plt.tight_layout()
plt.show()
````

### 2. "Shifting" A Matrix:-
As  with  vectors,  it  is  not  formally  possible  to  add  a  scalar  to  a  matrix,  as  in  λ+A.Python  allows  such  an  operation  (e.g.,  3+np.eye(2)),  which  involves  broadcast-adding  the  scalar  to  each  element  of  the  matrix.  That  is  a  convenient  computation,but it is not formally a linear algebra operation.But  there  is  a  linear-algebra  way  to  add  a  scalar  to  a  square  matrix,  and  that  iscalled shifting a matrix. It works by adding a constant value to the diagonal, which isimplemented by adding a scalar multiplied identity matrix:
<b>A+λI</b>
\begin{bmatrix}
4 & 5 & 1 \\
0 & 1 & 11 \\
4 & 9 & 7
\end{bmatrix} + 6
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}=
\begin{bmatrix}
10 & 5 & 1 \\
0 & 7 & 11 \\
4 & 9 & 13 \\
\end{bmatrix}

### “Shifting”  a  matrix  has  two  primary  (extremely  important!)  applications:<br> 1. It  is  the mechanism of finding the eigenvalues of a matrix. <br>2. It is the mechanism of regularizing matrices when fitting models to data.

````python
A = np.array([[4,5,1],[0,1,11],[4,9,7]])
lambda_scalar = 6
print(A + lambda_scalar)
print(A + lambda_scalar*np.eye(len(A)))
````

## 4. Standard Multiplication
Rather  than  operating  element-wise,  standard  matrix  multiplication  operates row/column-wise.  In  fact,  standard  matrix multiplication  reduces  to  a  systematic collection  of  dot  products  between  rows  of  one  matrix  and  columns  of  the  other matrix.This  form  of  multiplication  is  formally  simply  called  <i>matrix  multiplication</i>.

### You know that matrix sizes are written out as M × N — rows by columns. Two matrices  multiplying  each  other  can  have  different  sizes,  so  let’s  refer  to  the  size  of  the second  matrix  as  N×K.  When  we  write  out  the  two  multiplicand  matrices  with their  sizes  underneath,  we  can  refer  to  the  “inner”  dimensions  N  and  the  “outer” dimensions M and K. Here’s the important point: matrix multiplication is valid only when the “inner” dimensions match, and the size of the product matrix is defined by the “outer” dimensions. i.e <br><b><i><u>Matrix Multiplication can happen iff "No. of columns of first Matrix" = "No. of Rows of second Matrix". <br><br> The resultant matrix is of the shape "Rows of matrix 1 x Columns of matrix 2" </b></i></u>


The  reason  why  matrix  multiplication  is  valid  only  if  the  number  of  columns  in  the left matrix matches the number of rows in the right matrix is that the (i,j)$^{th}$ element in the product matrix is the dot product between the i$^{th}$ row of the left matrix and the j$^{th}$ column in the right matrix.

### Note   the   notation:   Hadamard   multiplication   is   indicated   using   a   dotted-circle (A⊙B)  whereas  matrix  multiplication  is  indicated  as  two  matrices  side-by-side without any symbol between them (AB).

### Example of matrix multiplication. Parentheses added to facilitate visual grouping.

## How  do  you  interpret  matrix  multiplication?

Remember  that  the  dot  product  is  anumber  that  encodes  the  relationship  between  two  vectors. So,  the  result  of matrix multiplication  is  a  matrix  that  stores  all  the  pairwise  linear  relationships  between rows  of  the  left  matrix  and  columns  of  the  right  matrix.  That  is  a  beautiful  thing,and is the basis for computing covariance and correlation matrices, the general linear model (used in statistics analyses including ANOVAs and regressions), singular-value decomposition, and countless other applications.

## Matrix-Vector Multiplication
1. A matrix can be right-multiplied by a column vector but not a row vector, and it can  be  left-multiplied  by  a  row  vector  but  not  a column vector.  In  other  words,Av and v$^T$A are valid, but Av$^T$ and vA are invalid. That is clear from inspecting matrix sizes: an M×N matrix can be pre-multiplied by  a  1 ×M  matrix  (a.k.a.  a  row  vector)  or  post-multiplied  by  an  N× 1  matrix(a.k.a. a column vector).
2. The result of matrix-vector multiplication is always a vector, and the orientation of that vector depends on the orientation of the multiplicand vector: pre-multiplying a matrix by a row vector produces another row vector, while post-multiplying a  matrix  by  a  column vector  produces  another  column  vector.  Again,  this  is obvious when you think about matrix sizes, but it’s worth pointing out.

### Revisiting Linear Weighted Combinations 
In the previous chapter, we calculated linear weighted combinations by having separate  scalars  and  vectors,  and  then  multiplying  them  individually. 
A  better,  more  compact,  and  more  scalable  method  for  computing  linear weighted combinations is to <i>put the individual vectors into a matrix, and put the weights into corresponding elements of a vector. Then multiply. </i>
#### Example
$$
\left. 4 \right.
\begin{bmatrix}
3 \\
0 \\
6
\end{bmatrix} + 3
\begin{bmatrix}
1\\
2\\
5
\end{bmatrix} =>
\begin{bmatrix}
3 & 1\\
0 & 2\\
6 & 5
\end{bmatrix}\begin{bmatrix}
4\\
3
\end{bmatrix}
$$

The  key  insight  is  that  each  element  in  the  vector scalar  multiplies  the  corresponding  column  in  the  matrix,  and  then  the weighted column vectors are summed to obtain the product.

### Geometric Transforms
When  we  think  of  a  vector  as  a  geometric  line,  then  matrix-vector  multiplication becomes  a  way  of  rotating  and  scaling  that  vector  (remember  that  scalar-vector multiplication can scale but not rotate).

### Matrix Operations: Transpose
The  principle  is the  same  with  matrices:  swap  the  rows  and  columns.  And  just  like  with  vectors,  the transpose  is  indicated  with  a  superscripted  $^T$  (thus,  C$^T$  is  the  transpose  of  C).  And double-transposing a matrix returns the original matrix (C$^{TT}$ = C).

#### Equation 5-2. Definition of the transpose operation
a$^{T}$$_{i,j}$ = a$_{j,i}$

#### Example
$$
\begin{bmatrix}
3 & 0 & 4\\
9 & 8 & 3
\end{bmatrix}^T =
\begin{bmatrix}
3 & 9 \\
0 & 8 \\
4 & 3
\end{bmatrix}
$$

````python
A = np.array([[3,0,4],[9,8,3]])
print(A)
print(A.shape)
print('**'*50)
A_transpose_method = A.T
print(A_transpose_method)
print(A_transpose_method.shape)
print('**'*50)
A_transpose_function = np.transpose(A) #Numpy function
print(A_transpose_function)
print(A_transpose_function.shape)
````

### Matrix Operations: LIVE EVIL (Order of Operations)
LIVE EVIL is a palindrome and  a  cute  mnemonic  for  remembering  how  transposing affects  the  order  of  multiplied  matrices.  Basically,  the  rule  is  that  <b><i><u>the  transpose  of multiplied matrices is the same as the individual matrices transposed and multiplied,but  reversed  in  order.</b></i></u><br>In  Equation  5-3,  L,  I,  V,  and  E  are  all  matrices,  and  you  can assume that their sizes match to make multiplication valid.

#### Equation 5-3. Example of the LIVE EVIL rule

(LIVE)$^T$ = E$^T$ V$^T$ I$^T$ L$^T$

### Symmetric Matrices
A matrix is said to be symmetric when  the  corresponding rows  and  columns  are  equal.And  that  means  that  when  you  swap  the  rows  and columns,  nothing  happens  to  the  matrix.  And  that  in  turn  means  that  a  symmetric matrix equals its transpose. In math terms, a matrix A is symmetric if A$^T$ = A. 
#### Equation 5-4. A symmetric matrix; note that each row equals its corresponding column
$$
\begin{bmatrix}
a & e & f & g \\
e & b & h & i \\
f & h & c & j \\
g & i & j & d 
\end{bmatrix}
$$

### Creating Symmetric Matrices from Nonsymmetric Matrices
Multiplying any  matrix even  a  nonsquare  and nonsymmetric  matrix by  its  transpose  will  produce  a  square  symmetric  matrix.  In other  words,A$^T$A  is   square  symmetric,  as    is   AA$^T$.

### Question 5.3:- Expand  this  code  to  extract  the  other  four  5 × 5  blocks.  Then  create  a  new  matrix with those blocks reorganized according to Figure 5-7.


````python
# Create the 10x10 matrix
matrix = np.arange(0, 100).reshape(10, 10)

# Partition the matrix into 4 submatrices
sub_matrices = [
    matrix[0:5, 0:5],  # Block 1
    matrix[0:5, 5:10], # Block 2
    matrix[5:10, 0:5], # Block 3
    matrix[5:10, 5:10] # Block 4
]

# Rearrange the blocks
rearranged_matrix = np.block([
    [sub_matrices[3], sub_matrices[2]],
    [sub_matrices[1], sub_matrices[0]]
])

# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Function to plot a matrix with annotations
def plot_matrix(ax, matrix, title):
    cmap = plt.get_cmap('gray')
    cax = ax.matshow(matrix, cmap=cmap, vmin=0, vmax=100)
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            # Use white text for darker backgrounds, black text for lighter backgrounds
            text_color = 'white' 
            ax.text(j, i, str(matrix[i, j]), va='center', ha='center', color=text_color)
    ax.set_title(title)
    # Draw dashed white lines to partition the blocks
    for i in range(1, 2):
        ax.axhline(i*5 - 0.5, color='white', linestyle='dashed', linewidth=1)
        ax.axvline(i*5 - 0.5, color='white', linestyle='dashed', linewidth=1)

# Plot the original matrix
plot_matrix(axs[0], matrix, 'Original Matrix with Partitions')

# Plot the rearranged matrix
plot_matrix(axs[1], rearranged_matrix, 'Rearranged Submatrices')

# Show the plot
plt.tight_layout()
plt.show()
````

### Question 5-8:- In  this  exercise,  you  will  write  a  Python  function  that  checks  whether  a  matrix  is symmetric.  It  should take  a  matrix  as  input,  and  should  output  a  boolean  True  if the  matrix  is  symmetric  or  False  if  the  matrix  is  non-symmetric.  Keep  in  mind  that small computer rounding/precision errors can make “equal” matrices appear unequal.

````python
def check_symmetric_matirx(matrix):
    rows, cols = matrix.shape
    # Check symmetric condition: matrix[i, j] should equal matrix[j, i] for all i, j
    for i in range(rows):
        for j in range(cols):
            if matrix[i, j] != matrix[j, i]:
                return False
    return True

def create_symmetric_matrix(matrix):
    # Ensure the matrix is square for symmetry operation
    rows, cols = matrix.shape
    if rows != cols:
        raise ValueError("Matrix must be square to make it symmetric.")

    # Iterate through upper triangle (excluding diagonal) to adjust values
    for i in range(rows):
        for j in range(i + 1, cols):  # Only need to adjust upper triangle
            if matrix[i, j] != matrix[j, i]:
                # Set symmetric elements
                matrix[i, j] = matrix[j, i]

    return matrix

matrix = np.random.randn(3,3)
is_symmetric = check_symmetric_matirx(matrix)
print(is_symmetric)
if not is_symmetric:
    symmetric_matrix = create_symmetric_matrix(matrix)
    is_symmetric = check_symmetric_matirx(matrix)
    print(is_symmetric)
    print(symmetric_matrix)
````
